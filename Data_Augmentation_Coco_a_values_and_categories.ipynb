# Data Augmentation Coco-a values and categories

with Back Translation "en" - "de" - "en"

```python
import os
from mtranslate import translate
```

```python
dir = "./train_data/1.4_5_6/coco_a_input/"
dir_out = "./train_data/augm_coco_a/augm_coco_a_input/"
print("1")
```

```python
data = []
for f in os.listdir(dir):
    data.append(f)
    
print("done ",data)
```

Attributes:

```python
with open (dir+data[1], "r") as d_a:
    print("dev", data[1])
    dev_attr = d_a.readlines()
with open(dir_out+"dev.attr", "a") as f:
    for line in dev_attr:
        f.write(line)
        f.write(line)

with open (dir+data[6], "r") as tr_a:
    print("train", data[6])
    train_attr = tr_a.readlines()
with open(dir_out+"train.attr", "a") as f:
    for line in train_attr:
        f.write(line)
        f.write(line)

with open (dir+data[4], "r") as tst_a:
    print("test", data[4])
    tst_attr = tst_a.readlines()
with open(dir_out+"test.attr", "a") as f:
    for line in tst_attr:
        f.write(line)
        f.write(line)
        
with open (dir+data[3], "r") as tst_a:
    tst_attr = tst_a.readlines()
    print("im", data[3])
with open(dir_out+"image_id.txt", "a") as f:
    for line in tst_attr:
        f.write(line)
        f.write(line)
        
with open (dir+data[8], "r") as tst_a:
    print("voc", data[8])
    tst_attr = tst_a.readlines()
with open(dir_out+"vocab.attr", "a") as f:
    for line in tst_attr:
        f.write(line) # just 1 time (simply copy the file)
        #f.write(line) 
        
print("Done")
```

Dev caprions:

```python
transl = []
with open (dir+data[2], "r") as d_c:
    dev_capt = d_c.readlines()
with open(dir_out+"dev.capt", "a",encoding='utf8') as f:
    for line in dev_capt:
        print(line)
        try:
            german_translation = translate(line, "de")
            en_translation = translate(german_translation, "en")
        except:
            en_translation = ""
            line1 = line.split("</s>")
            for el in line1:
                german_translation = translate(el, "de")
                en_tr = translate(german_translation, "en")
                en_translation+=en_tr
                en_translation+=" </s> "
                #print()
            #print("! ",en_translation)
        f.write(line)        
        f.write(en_translation)
        f.write("\n")

print("Done")

```

Train captions

```python
transl = []
with open (dir+data[7], "r") as t_c:
    train_capt = t_c.readlines()
with open(dir_out+"train.capt", "a", encoding='utf8') as f:
    for line in train_capt:
        print(line)
        try:
            german_translation = translate(line, "de")
            en_translation = translate(german_translation, "en")
        except:
            en_translation = ""
            line1 = line.split("</s>")
            print(line1)
            for el in line1:
                print(el)
                german_translation = translate(el, "de")
                en_tr = translate(german_translation, "en")
                en_translation+=en_tr
                en_translation+=" </s> "
                #print()
            #print("! ",en_translation)
        f.write(line)        
        f.write(en_translation)
        f.write("\n")

print("Done")
```

Test Captions

```python
transl = []
with open (dir+data[5], "r") as tst_c:
    tst_capt = tst_c.readlines()
with open(dir_out+"test.capt", "a", encoding='utf8') as f:
    for line in tst_capt:
        print(line)
        try:
            german_translation = translate(line, "de")
            en_translation = translate(german_translation, "en")
        except:
            en_translation = ""
            line1 = line.split("</s>")
            for el in line1:
                german_translation = translate(el, "de")
                en_tr = translate(german_translation, "en")
                en_translation+=en_tr
                en_translation+=" </s> "
                #print()
            #print("! ",en_translation)
        f.write(line)        
        f.write(en_translation)
        f.write("\n")

print("Done")
```

Vocab captions

```python
vocab = []

with open (dir+data[2], "r") as d_c:
    dev_capt = d_c.readlines()
with open (dir+data[7], "r") as t_c:
    train_capt = t_c.readlines()
with open (dir+data[9], "r") as tst_c:
    tsts_capt = tst_c.readlines()

for el in dev_capt:
    for e in el.split():
        if e.lower() not in vocab:
            vocab.append(e.lower())
            vocab.append("\n")
for el in train_capt:
    for e in el.split():
        if e.lower() not in vocab:
            vocab.append(e.lower())
            vocab.append("\n")
for el in tsts_capt:
    for e in el.split():
        if e.lower() not in vocab:
            vocab.append(e.lower())
            vocab.append("\n")
print(vocab)
with open(dir_out+"vocab.capt", "a", encoding='utf8') as f:
    for el in vocab:
        f.write(el)
print("Done")
```
